```python
# normalize the features using mean normalization
data_setNorm = (data_set - data_set.mean())/data_set.std()
data_setNorm.head()
```

```python
dataTest = data_setNorm.iloc[17300:,:] # 4400 test examples
dataTrain = data_setNorm.iloc[0:17300,:] # 17300 training examples
```

```python
X_dataTrain = dataTrain.iloc[:,0:11] #Features
```

```python
Y_dataTrain = dataTrain.iloc[:,11:12] #Lables
```

```python
m = len(X_dataTrain)
print(m)
# Size of training data
```

```python
ones = pd.Series(1,np.arange(m)) # Creating araay of ones 
```

```python
X_dataTrain = pd.concat([ones,X_dataTrain],axis=1)
print(X_dataTrain)
```

```python
theta = np.zeros([1,12])
print(theta)
```

```python
#set hyper parameters
alpha = 0.01
iters = 1000
```

Cost function for Linear Reggression 
<img src ="https://cdn-images-1.medium.com/max/800/1*VanG05Ab6yknqJ2bRGFzrQ.png">

```python
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))
```

```python
initCost = computeCost(X_dataTrain.values,Y_dataTrain.values,theta)
print("Initial Cost of the model is {}".format(initCost))
```

```python
#gradient descent
def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost
```

```python
#running The Gradient Descent
minTheta,cost = gradientDescent(X_dataTrain.values,Y_dataTrain.values,theta,iters,alpha)
print(minTheta)
```

```python
finalCost = computeCost(X_dataTrain.values,Y_dataTrain.values,minTheta)
print(finalCost)
```

```python
#plot the cost
fig, ax = pt.subplots()  
ax.plot(np.arange(iters), cost, 'r')  
ax.set_xlabel('Iterations')  
ax.set_ylabel('Cost')  
ax.set_title('Error vs. Training Epoch')  
```
